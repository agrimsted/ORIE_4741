```python
birthgroups = X[(X["birth.year"] > 0)]
birthgroups["age"] = 2019 - birthgroups["birth.year"]
print(birthgroups)
```

```python
age_group = []
sort_group = []
for i in birthgroups['age']:
    if i < 20 :
        age_group.append("17-19")
        sort_group.append(1)
    elif i < 30:
        age_group.append("20-29")
        sort_group.append(2)
    elif i < 40:
        age_group.append("30-39")
        sort_group.append(3)
    elif i < 50 :
        age_group.append("47-49")
        sort_group.append(4)
    elif i < 60:
        age_group.append("50-59")
        sort_group.append(5)
    elif i < 70:
        age_group.append("60-69")
        sort_group.append(6)
    elif i < 80 :
        age_group.append("70-79")
        sort_group.append(7)
    elif i < 90:
        age_group.append("80-89")
        sort_group.append(8)
    elif i < 100:
        age_group.append("90-99")
        sort_group.append(9)
    elif i < 110 :
        age_group.append("100-109")
        sort_group.append(10)
    elif i < 120:
        age_group.append("110-119")
        sort_group.append(11)
    elif i < 130:
        age_group.append("120-129")
        sort_group.append(12)
    elif i < 140:
        age_group.append("130-139")
        sort_group.append(13)
    elif i < 150:
        age_group.append("140-149")
        sort_group.append(14)
    elif i < 160:
        age_group.append("150-159")
        sort_group.append(15)
        
age_group
```

```python
birthgroups["age_groups"] = age_group
birthgroups["sort_groups"] = sort_group
birthgroups
```

```python
plt.figure(6)
birthgroups.sort_values(by=['sort_groups'])
fig, ax = plt.subplots()
sns.boxplot(x="sort_groups", y='tripduration',data=birthgroups)
ax.set_ylim(0,3500)
plt.title("Tripduration over age groups")
plt.xlabel("Age Groups")
plt.xticks(np.arange(13), ("17-19","20-29","30-39","40-49","50-59","60-69","70-79","80-89","90-99","100-109","110-119",
                       "120-129","130-139"),rotation='vertical')
fname="Trip_Duration_age_group.pdf"
plt.savefig(fname)
plt.show()
```
